```python
import pandas as pd

# 将列拆分为两列，并处理空值
df[['col1', 'col2']] = df['会员位置(GPS)'].str.split(' ', expand=True).fillna(0).astype(float)

# 打印拆分后的 DataFrame
print(df)
```

```python
import pandas as pd

# 将列拆分为两列，并处理空值
split_data = df['会员位置(GPS)'].str.split(' ', expand=True)
df['col1'] = pd.to_numeric(split_data[0], errors='coerce')
df['col2'] = pd.to_numeric(split_data[1], errors='coerce')

# 打印拆分后的 DataFrame
print(df)
```

```python
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 使用3σ原则识别离群点
mean = df['col1'].mean()
std = df['col1'].std()
threshold = 3 * std
outliers = df[(df['col1'] > mean + threshold) | (df['col1'] < mean - threshold)]

# 打印离群点
print("Outliers:")
print(outliers[['会员编号','col1']])

# 使用3σ原则识别离群点
mean = df['col2'].mean()
std = df['col2'].std()
threshold = 3 * std
outliers = df[(df['col2'] > mean + threshold) | (df['col2'] < mean - threshold)]

# 打印离群点
print("Outliers:")
print(outliers[['会员编号','col2']])
```
