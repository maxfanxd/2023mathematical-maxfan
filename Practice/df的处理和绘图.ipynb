```python
### 三种函数的参数
# 默认值
def f(a,b=1):
    print(b)
f(1,2)
f(1,1)
f(2)
f(a=2)

print ('')

# 多个参数
def f(* args):
    for i in args:
        print(i)
f(1,2,3)
print('')
f(4)

# 字典（默认key是某种参数名）
def f(** dict):
    for i,j in dict.items():
        print(i)
        print(j)

f(country = 'spain', capital = 'madrid', population = 123456)
```

# 探索性数据分析
首先我们考虑频率计数

假设 75% 的值是 Q3，25% 的值是 Q1


异常值小于 Q1 - 1.5(Q3-Q1) 且大于 Q3 + 1.5(Q3-Q1)。 (Q3-Q1) = IQR

- **什么是分位数？**

1,4,5,6,8,9,11,12,13,14,15,16,17
中位数是位于序列中间的数字。 在这种情况下，它将是 11

下四分位数是最小数字和中位数之间的中位数，即 1 到 11 之间的中位数，即 6

在上四分位数中，您可以找到中位数和最大数字之间的中位数，即 11 到 17 之间，根据上述问题，该中位数为 14

```python
data.boxplot(column='Attack',by = 'Legendary')

# 例子：比较传奇与非传奇神奇宝贝的攻击力
# 顶部黑线为最大
# 顶部的蓝线是 75%
# 绿线是中位数 (50%)
# 底部的蓝线是 25%
# 底部黑线是最小值
```

```python
### 使用melt()重塑数据
data_new = data.head()    # I only take 5 rows into new data
data_new
```

```python
# melt的过程
# id_vars = what we do not wish to melt
# value_vars = what we want to melt
melted = pd.melt(frame=data_new,id_vars = 'Name', value_vars= ['Attack','Defense'])
melted
```

```python
# pivot，即melt的reverse过程
# Index is name
# I want to make that columns are variable
# Finally values in columns are value
melted.pivot(index = 'Name', columns = 'variable',values='value')
```

```python
# 这里应该会有缺失值的问题，我们来看看
melted[:9].pivot(index = 'Name', columns = 'variable',values='value')

# 没错，可以看到出现了NaN值
```

## 连接(concatenating)数据

```python
# 行方向合并，主要参数是axis=0
data1 = data.head()
data2= data.tail()
conc_data_row = pd.concat([data1,data2],axis =0,ignore_index =True) # axis = 0 : adds dataframes in row
conc_data_row
```

```python
# 列方向上的合并
data1 = data['Attack'].head()
data2= data['Defense'].head()
conc_data_col = pd.concat([data1,data2],axis =1) # axis = 1 : adds dataframes in column
conc_data_col
```
